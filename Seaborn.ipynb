```python
#joint grid
g=sns.JointGrid(x='total_bill',y='tip',data=df)
g=sns.jointplot(x='total_bill',y='tip',data=df)
g=g.plot(sns.regplot,sns.distplot)
```

```python
import seaborn as sns
lip=sns.load_dataset('tips')
g=sns.JointGrid(x='total_bill',y="tip",data=lip)
g=sns.jointplot(x='total_bill',y="tip",data=lip,kind='hex')
g=g.plot(sns.regplot,sns.distplot)
```

```python
g=sns.jointplot(x='total_bill',y="tip",data=lip,kind='reg')
g=g.plot(sns.regplot,sns.distplot)
```
